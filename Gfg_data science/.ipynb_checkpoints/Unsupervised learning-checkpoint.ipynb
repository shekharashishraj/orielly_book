```python
import pandas as pd
import numpy as np
import sklearn 
from sklearn.cluster import KMeans
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/ashishrajshekhar/Desktop/Tution notes and class/GL_STATISTICS/Wholesale customers data.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_df = scaler.fit_transform(df)
pd.DataFrame(scaled_df,columns=df.columns).describe()

#scaled_df
```

```python
scaled_df.shape
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init = 'random', n_init = 'auto', random_state = 12)
    kmeans.fit(df)
    inertia.append((i,kmeans.inertia_,))
print(inertia)
plt.plot([w[0] for w in inertia],[w[1] for w in inertia], marker="X")
```

```python

wcss=[]
for  i in range(1,11):
    kmeans=KMeans(n_clusters=i ,init = 'random', n_init = 'auto',random_state=10)
    kmeans.fit(df)
    wcss.append((i,kmeans.inertia_))
print(wcss)
plt.plot([w[0] for w in inertia],[w[1] for w in wcss], marker="X")
```

```python
wcss
```

```python

cluster = KMeans(n_clusters=2,random_state =13)
cluster.fit(df)
df['cluster']= cluster.labels_
sns.countplot(data= df, x='cluster')
```
